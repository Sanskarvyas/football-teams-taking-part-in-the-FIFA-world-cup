```python
import pandas as pd
import numpy as np

#Importing and clearing the dataset.
'''
In this code we have imported the dataset which details the statistics of the football teams taking part in the FIFA world cup. 
After importing this dataset, we have proceed to organize the data_Set on the basis of classifications: Group and Team. 
We proceed to declutter and clean the data by dropping statistics that are irrelavant to our narrative
'''

group_stats = pd.read_csv("group_stats.csv") #importing the dataset
group_stats = group_stats.pivot_table(index = ["group","team"]) #index has been designated
group_stats = group_stats.drop(columns= ["Unnamed: 0","exp_goal_conceded","exp_goal_difference","exp_goal_difference_per_90","expected_goal_scored"]) #clearing irrelavant data
group_stats


```

```python
# 1 Which team in each group has the highest goal difference?

group_stats['goal_difference'] = group_stats['goals_scored'] - group_stats['goals_against'] #defining goal difference
highest_goal_diff_teams = group_stats.groupby('group')['goal_difference'].idxmax().apply(lambda x: x[1]) #indentifying the highest goal-scoring team in each group 
                                                                                                         

for group, team in highest_goal_diff_teams.items():
    print(f"The team with the highest goal difference in Group {group} is {team}")

'''
The lambda function is extracting the second element of each group: which is team in accordance to our query 
'''
```

```python
# 2 What is the total number of goals scored in each group?

total_goals_by_group = group_stats.groupby('group')['goals_scored'].sum() #calculating the total number of goals scored in each group

for group, total_goals in total_goals_by_group.items():
    print(f"The total number of goals scored in Group {group} is {total_goals}")
```

```python
# 3 Which team has the highest number of wins in their respective groups?
wins_by_group_team = group_stats.groupby(['group', 'team'])['wins'].sum() #identifying the most winning team in each group
teams_with_most_wins = wins_by_group_team.groupby('group').idxmax().apply(lambda x: x[1]) 

for group, team in teams_with_most_wins.items():
     print(f"The team that has the highest number of wins in their respective group {group} is {team}")
```

```python
# 4 How many matches ended in a draw in each group?
draws_by_group = group_stats[group_stats['draws'] == 1].groupby('group').size() 

for group, num_draws in draws_by_group.items():
    
    print(f"The number of matches that ended in a draw in Group {group} is {num_draws}")
```

```python
# 5 Which team had the worst performance in terms of losses and goals against amongst ALL the groups?

'''
To answer this questions we must calculate a performance score that assess teams on the basis of a few performance facets
here, the worst team would be one that has conceeded the most goals and lost the most
'''

# We calculate a performance score: 
group_stats['performance_score'] = group_stats['losses'] + group_stats['goals_against']

# We Find the team with the worst performance
worst_team = group_stats.loc[group_stats['performance_score'].idxmax()].name[1]

print(f"The team with the worst performance in terms of losses and goals against is: {worst_team}")
```

```python
# 6 Which team had the worst performance in terms of losses and goals against within their own respective groups?

# We calculate performance score
group_stats['performance_score'] = group_stats['losses'] + group_stats['goals_against']

''' 
From the previous question, to find the worst performing team in each group we use the lambda function and 
sequence our analysis on the basis of group.
'''

# We find the team with the worst performance in each group
worst_teams_by_group = group_stats.groupby('group')['performance_score'].idxmax().apply(lambda x: x[1])

for group, team in worst_teams_by_group.items():
    
    print(f"The team with the the worst performance in Group {group} is {team}")
```

```python
# 7 which team had the best performance in terms of victories and goals scored within their own respective groups? 

'''
To answer this questions we must calculate a performance score that assess teams on the basis of a few performance facets
here, the best team would be one that has scored the most goals and won the most
'''

# We calculate performance score
group_stats['performance_score'] = group_stats['wins'] + group_stats['goals_scored']

# We find the team with the best performance in each group
best_teams_by_group = group_stats.groupby('group')['performance_score'].idxmax().apply(lambda x: x[1])

for group, team in best_teams_by_group.items():
    
    print(f"The team with the the best performance in Group {group} is {team}")
```

```python
# 8 Which team had the best performance in terms of wins and goals scored amongst ALL the groups?

'''
To answer this questions we must calculate a performance score that assess teams on the basis of a few performance facets
here, the best team would be one that has scored the most goals and won the most.
'''

# We calculate a performance score: 
group_stats['performance_score'] = group_stats['wins'] + group_stats['goals_scored']

# We Find the team with the worst performance
best_team = group_stats.loc[group_stats['performance_score'].idxmax()].name[1]

print(f"The team with the best performance in terms of wins and goals scored amongst all groups is: {best_team}")
```
