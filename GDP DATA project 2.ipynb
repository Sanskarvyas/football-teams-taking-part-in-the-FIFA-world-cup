```python
#List of countries GDP from 1960 to 2022
import matplotlib.pyplot as plt
countries_to_analyze = ['India', 'United States', 'China', 'Qatar', 'United Kingdom']

selected_countries_df = df[df['Country Name'].isin(countries_to_analyze)]

pivot_df = selected_countries_df.pivot(index='Year', columns='Country Name', values='Value')

for country in countries_to_analyze:
    plt.plot(pivot_df.index, pivot_df[country], label=country)
plt.title('GDP Trends for Selected Countries (1960-2017)', loc='right', pad=20)
plt.xlabel('Year')
plt.ylabel('GDP Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Comparision of GDP trends in South Asia 

import matplotlib.pyplot as plt

#South Asian countries to compare

South_Asian_countries_compare = { 'South Asia': ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Sri Lanka', 'Pakistan'],}

selected_regions_df = df[df['Country Name'].isin([country for countries in South_Asian_countries_compare.values() for country in countries])]

pivot_df = selected_regions_df.pivot(index = 'Year', columns = 'Country Name', values = 'Value')

#plotting with specified colours
plt.figure(figsize=(8, 6))
for region, countries in South_Asian_countries_compare.items():
    for country in countries:
        color = {
            'India': 'blue',
            'Afghanistan': 'red',
            'Pakistan': 'black',
            'Nepal': 'orange',
        }.get(country, 'gray')  
        plt.plot(pivot_df.index, pivot_df[country], label=country, color=color)
           
plt.title('GDP TRENDS FOR SOUTH ASIAN COUNTRIES FROM (1960-2017)', loc = 'right', pad = 20)
plt.xlabel('Year')
plt.ylabel('GDP Value in billions of dollars')
plt.legend()
plt.show()
                                          
            
```

```python
# Comparison of GDP Tends in Western and Eastern Europe


# Western and Eastern Europe countries with country codes
Europe_compare = {
    'Western Europe': ['AUT', 'BEL', 'CHE', 'DEU', 'DNK', 'ESP', 'FIN', 'FRA', 'GBR', 'GRC', 'IRL', 'ITA', 'LUX', 'NLD', 'NOR', 'PRT', 'SWE'],
    'Eastern Europe': ['ALB', 'BIH', 'BGR', 'HRV', 'CZE', 'EST', 'HUN', 'LTU', 'LVA', 'MKD', 'MNE', 'POL', 'ROU', 'SRB', 'SVK', 'SVN', 'TUR', 'UKR']
}

# Setting up the subplots that I have decided on: Western and Eastern Europe
fig, axs = plt.subplots(2, 1, figsize=(12, 16), sharex=True)

# Iterate through each region
for i, (region, countries) in enumerate(Europe_compare.items()):
    selected_regions_df = df[df['Country Code'].isin(countries)]
    pivot_df = selected_regions_df.pivot(index='Year', columns='Country Code', values='Value')

    # Plotting
    for country in countries:
        axs[i].plot(pivot_df.index, pivot_df[country] / 1e9, label=country)  # Divided by 1e9 to convert to billions

    # Title and labels for each subplot
    axs[i].set_title(f'GDP Trends for {region} Countries (1960-2017)')
    axs[i].set_ylabel('GDP Value (Billions of Dollars)')
    axs[i].legend()

# Common x-axis label
axs[1].set_xlabel('Year')
plt.show()

```

```python
import matplotlib.pyplot as plt

# North Africa, Central Africa, and Southern Africa countries with country codes
Africa_compare = {
    'North Africa': ['DZA', 'EGY', 'LBY', 'MAR', 'SDN', 'TUN'],
    'Central Africa': ['AGO', 'CMR', 'CAF', 'TCD', 'COD', 'COG', 'GAB', 'STP'],
    'Southern Africa': ['BWA', 'COM', 'LSO', 'MUS', 'NAM', 'SWZ', 'ZAF', 'ZMB', 'ZWE'],
}

# Loop 
for region, countries in Africa_compare.items():
    # DataFrame for the specified countries
    selected_regions_df = df[df['Country Code'].isin(countries)]

    # DataFrame for the region
    pivot_df = selected_regions_df.pivot(index='Year', columns='Country Code', values='Value')

    # Plotting
    plt.figure(figsize=(12, 8))
    for country in countries:
        plt.plot(pivot_df.index, pivot_df[country] / 1e9, label=country)  # Divide by 1e9 to convert to billions

    # x-axis 
    plt.xticks(range(1960, 2023, 5))

    # y-axis is auto-generated

    # Customize the title
    plt.title(f'GDP Trends for {region} Countries (1960-2017)', loc='right', pad=20)

    # Adjust spacing to avoid overlap of y-axis labels
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

    plt.xlabel('Year')
    plt.ylabel('GDP Value (Billions of Dollars)')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=2)
    plt.show()
```
